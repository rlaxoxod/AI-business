## 종료시기 낮은 co2를 배출하기 위한 브랜드별 최적조건

* 종료시기 브랜드별 co2 최저값

```python
# 최솟값을 가질 때의 조건 추출
brand_min_co2 = df.groupby('brand')['future_co2'].min()

# 최솟값을 가질 때의 조건 추출
min_co2_conditions = df[df['future_co2'].isin(brand_min_co2.values)]
```

```python
min_co2_conditions.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data = min_co2_conditions
variables = ['vehicle_class', 'transmission', 'fuel_type']
for var in variables:
    # 더미 변수 추출
    dummy_columns = [column for column in data.columns if column.startswith(var + '_')]
    
    # 변수 복구
    data[var] = data[dummy_columns].idxmax(axis=1).str.replace(var + '_', '')

    # 더미 변수 삭제
    data.drop(dummy_columns, axis=1, inplace=True) 
```

```python
data.columns
```

```python
data = data[['brand', 'period','hybrid', 'engine_size', 'cylinders', 'gears', 
             'vehicle_class', 'transmission', 'fuel_type', 'fuel_cons_comb', 'future_co2']]
```

## 2 by 2 matrix

* 기준: 종료시기까지 남은 기간, (종료시기-2023)_co2

```python
df3 = pd.read_csv('classify.csv')
```

```python
df3
```

```python
df3.columns
```

```python
train = df3[['remain_period', '2023_co2', 'reduce_gap']]
```

```python
from pycaret.clustering import *
cl = setup(data=train,
          normalize=True,
          session_id=123)
```

```python
kmeans = create_model('kmeans')
```

```python
predictions = assign_model(kmeans)
```

```python
predict_label = predictions
```

```python
predict_label
```

```python
predict_label.to_csv('predict_label.csv')
```

```python
result = predict_label.groupby('Cluster').agg({
    'remain_period' : ['mean'],
    '2023_co2' : ['mean'],
    'reduce_gap' : ['mean']
})
```

```python
result
```

```python
result.columns
```

```python
result.columns = result.columns.droplevel(1)
```

```python
result.index
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_result = pd.DataFrame(scaler.fit_transform(result), columns=result.columns, index=result.index)

plt.figure(figsize=(8, 6))
for column in scaled_result.columns:
    plt.plot(scaled_result.index, scaled_result[column], marker='o', linestyle='-', linewidth=2, label=column)
#plt.xlabel('Cluster')
plt.ylabel('Value')
plt.title('Snake Plot')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=len(scaled_result.columns))
plt.show()
```

* cluster distribution

```python
# predict_label cluster rename
predict_label['Cluster'] = predict_label['Cluster'].replace({'Cluster 0': 'Latecomer',
                                                             'Cluster 1': 'Eco Friendly',
                                                             'Cluster 2': 'Realistic',
                                                            'Cluster 3' : 'Not Over Yet'})
```

```python
# brand name
predict_label['brand'] = df3['brand']
```

```python
cluster_result = predict_label[['brand', 'Cluster']]
```

```python
def print_brands_by_cluster(cluster_df):
    cluster = cluster_df['Cluster'].iloc[0]  # 현재 cluster 번호
    brands = cluster_df['brand'].unique()    # 현재 cluster에 속하는 brand들의 유일한 값들
    #print(f"Cluster {cluster}: {brands}")
    return brands
# Cluster 별 brand들 출력
final_result = cluster_result.groupby('Cluster').apply(print_brands_by_cluster)
```
